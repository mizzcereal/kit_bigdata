데이터 전처리 : 분석에 적합하게 데이터를 가공하는 작업

```python
import pandas as pd
exam = pd.read_csv('./exam.csv')
exam
```

```python
# 1반 학생들만 추출
exam.query('nclass == 1')
```

```python
# 수학 점수가 50점 이상인 학생 출력
exam.query('math > 50')
```

```python
exam.query('english <= 80')
```

```python
# 1반이 아닌 학생 출력
exam.query('nclass != 1')
```

&  : 조건 모두 만족하는 것이 True
| : 조건 중 하나만 만족해도 True

```python
# 여러 조건을 충족하는 행 추춣

exam.query('nclass == 1 & math >= 50')
```

```python
exam.query('math >= 90 | english >= 90')
```

목록에 해당하는 행 추출은 |을 이용해 여러 조건을 나열하면 된다.
in[]을 사용해 보다 간결하게 사용할 수 있다.

```python
exam.query('nclass == 1 | nclass == 3 | nclass ==5')
```

```python
exam.query('nclass in [1,3,5]')
```

```python
nclass1 = exam.query('nclass ==1')
nclass2 = exam.query('nclass ==2')

print(nclass1['math'].mean())
print(nclass2['math'].mean())
```
