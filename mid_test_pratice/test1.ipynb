6장 조건에 맞는 데이터 추출 혼자서 해보기


```python
# Q1. 자동차 배기량에 따라 고속도로 연비가 다른지 알아볼때 displ(배기량)이 4이하인 자동차와 5이상인 자동차 중 어떤 자동차의 hwy(고속도로 연비)
#평균이 높은지 조사

import pandas as pd
import numpy as np

mpg = pd.read_csv('./mpg.csv')

mpg4 = mpg.query('displ <= 4')
mpg5 = mpg.query('displ >= 5')

print(mpg4['hwy'].mean())
print(mpg5['hwy'].mean())
```

```python
#Q2. 자동자 제조 회사에 따라 도시연비가 어떻게 다른지 확인
#audi와 toyota 중 어느 manufacturer(제조회사)의 cty(도시연비) 평균이 더 높은지 조사

mpg_audi = mpg.query('manufacturer == "audi"')
mpg_toyota = mpg.query('manufacturer == "toyota"')

print(mpg_audi['cty'].mean())
print(mpg_toyota['cty'].mean())
```

```python
#Q3. 'chevrolet', 'ford', 'honda' 자동차의 hwy 평균을 알아보고조한다
# 세 회사의 데이터를 추출하고 hwy의 전체 평균을 구하라

mpg3 = mpg.query('manufacturer in ["chevrolet", "ford", "honda"]')
mpg3['hwy'].mean()
```

필요한 변수 출력 문제

```python
#Q1. mpg데이터는 11개 변수로 구성 이중 category, cty 변수만 추출해 새로운 데이터를 생성

mpg_new = mpg[['category', 'cty']]
mpg_new
```

```python
mpg_new_suv = mpg_new.query('category == "suv"')
mpg_new_compact = mpg_new.query('category == "compact"')
print(mpg_new_suv['cty'].mean())
print(mpg_new_compact['cty'].mean())
```

파생변수 만들기 .assign() 문제

```python
#Q1. mpg 데이터 복사본 만들고 cty와 hwy를 더한 합산 연비 변수를 축
mpg_new = mpg
mpg_new = mpg_new.assign(합산_연비 = mpg_new['cty'] + mpg_new['hwy'])
```

```python
#Q2. 합산 연비 변수를 2로 나눠 평균 연비 변수 추가
mpg_new = mpg_new.assign(평균_연비 = (mpg_new['합산_연비'])/2)
```

```python
#Q3. 평균 연비 변수가 가장 높은 자동차 3종 데이터 출력
mpg_new.sort_values('평균_연비', ascending=False).head(3)
```

```python
#Q4. 구문 하나로 합치기
mpg.assign(합산_연비 = lambda x: x['cty'] + x['hwy'],
           평균_연비 = lambda x: x["합산_연비"]/2) \
           .sort_values('평균_연비', ascending= False)\
           .head(3)
```

pandas 함수 조합

```python
# 제조회사별 suv 자동차의 도시 고속도로 합산 연비 평균 구해 내림차순 정렬 후 1~5위 출력 

mpg.query('category == "suv"')\
   .assign(total = (mpg['hwy'] + mpg['cty'])/2)\
   .groupby('manufacturer')\
   .agg(합산_연비_평균 = ('total', 'mean'))\
   .sort_values('합산_연비_평균', ascending=False)\
   .head()
   
```

```python
#Q1 mpg 데이터 category는 자동차 특징에 따라 일곱 종류로 분류 카테골별 cty 평균 구하기
mpg.groupby('category')\
   .agg(cty_mean = ('cty','mean'))
```

```python
#Q2. Q2을 cty평균이 높은 순으로 정렬
mpg.groupby('category')\
   .agg(cty_mean = ('cty','mean'))\
   .sort_values('cty_mean', ascending= False)
```

```python
#Q3. hwy평균이 가장 높은 회사 세 곳 출력

mpg.groupby('manufacturer')\
   .agg(hwy_mean = ('hwy', 'mean'))\
   .sort_values('hwy_mean', ascending=False)\
   .head(3)
```

```python
#Q4. 어떤 회사에서 compact 차종을 가장 많이 생산하늕 알아보라(내림차순 정렬)

mpg.query('category == "compact"')\
   .groupby('manufacturer')\
   .agg(compact_count = ('manufacturer', 'count'))\
   .sort_values('compact_count', ascending=False)
```
