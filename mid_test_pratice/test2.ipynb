```python
# mpg 데이터 복사본 만들고 합산 연비 변수 추가

import pandas as pd
import numpy as np

mpg = pd.read_csv('./mpg.csv')
mpg_new = mpg

mpg_new = mpg_new.assign(합산연비변수 = lambda x : x['cty'] + x['hwy'])

mpg_new
```

```python
# 평균 연비 변수도 추가하고 가장 높은 3종데이터
mpg_new = mpg_new.assign(평균연비변수 = mpg_new['합산연비변수']/2)
mpg_new.sort_values('평균연비변수', ascending=False).head(3)
```

```python
mpg.assign(합산연비변수 = lambda x : x['cty'] + x['hwy'],
           평균연비변수 = lambda x : x['합산연비변수']/2)\
           .sort_values('평균연비변수', ascending=False).head(3)
```

```python
#제조회사별 suv 자동차의 도시 및 고소도록 ㅎ합산 연비 평균 구하고 내림차순 정렬 후 1~ 5위 출력
mpg.query("category == 'suv'")\
    .assign(합산연비 = mpg['cty'] + mpg['hwy'])\
   .groupby('manufacturer')\
   .agg(합산연비평균 = ('합산연비', 'mean'))\
   .sort_values('합산연비평균', ascending=False).head()
```

```python
# category는 차종 뷴리한것 category별 cty평균구하기

mpg.groupby('category')\
   .agg(cty평균 = ('cty', 'mean'))
```

```python
# 위의 구한것을 cty평균이 높은 순으로 정렬
mpg.groupby('category')\
   .agg(cty평균 = ('cty', 'mean'))\
   .sort_values('cty평균', ascending=False)
```

```python
# 회사별 hwy가 평균이 가장 높으 회사 3곳 출력

mpg.groupby('manufacturer')\
   .agg(hwy평균 = ('hwy', 'mean'))\
   .sort_values('hwy평균', ascending=False).head(3) 
```

```python
# 회사별 commpact 차종 수를 내림차순으로 정리
mpg.query('category == "compact"')\
   .groupby('manufacturer')\
   .agg(compact차종수 = ('manufacturer', 'count'))\
   .sort_values('compact차종수', ascending=False)
```

```python
# f1 데이터 프레임 만들기
fuel = pd.DataFrame({'fl' : ['c','d','e','p','r'],
                     '연료종류' : ['CNG', 'diesel', 'ethnaol E85', 'premiun', 'regular'],
                     "가격" : [2.35, 2.38, 2.11, 2.76, 2.22]})
fuel
```

```python
#mpg 데이터에서 fuel데이터에 있는 가격 변수 추가

mpg = pd.merge(mpg, fuel, how='left', on = 'fl')
mpg
```

```python
# midwest데이터에서 '전체인구 대비 미성년 인구 백분율 구하기

midwest = pd.read_csv('./midwest.csv')
midwest = midwest.assign(미성년인구백분율 = (midwest['poptotal'] - midwest['popadults'])/midwest['poptotal'] * 100)
midwest
```

```python
# 미성년 인구 백분율 가장 높은 상위 5개 지역의 백분율 출력
midwest.sort_values('미성년인구백분율', ascending=False).head()\
       [['county', '미성년인구백분율']]
```

```python
midwest = midwest.assign(등급 = np.where(midwest['미성년인구백분율'] >= 40, 'Large',
                             np.where(midwest['미성년인구백분율'] >= 30, "Middle", "Small")))
midwest.value_counts('등급')
```
